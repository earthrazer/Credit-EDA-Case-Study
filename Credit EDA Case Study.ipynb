#### We will write a function to plot four graph for Approved,Refused,canceled and unused

```python
def plot(x):
    fig, ((x1, x2), (x3, x4)) = plt.subplots(nrows=2,ncols=2, figsize=(15,10), sharey='all')
    fig.tight_layout(pad=10.0)

    plt.subplot(2,2,1)
    sns.countplot(data=approved,x=x,ax=x1)
    plt.title("Approved")
    plt.ylabel("Number of Loans")
    plt.xticks(rotation=90)

    plt.subplot(2,2,2)
    sns.countplot(data=refused,x=x,ax=x2)
    plt.title("Refused")
    plt.ylabel("Number of Loans")
    plt.xticks(rotation=90)

    plt.subplot(2,2,3)
    sns.countplot(data=canceled,x=x,ax=x3)
    plt.title("Canceled")
    plt.ylabel("Number of Loans")
    plt.xticks(rotation=90)

    plt.subplot(2,2,4)
    sns.countplot(data=unused,x=x,ax=x4)
    plt.title("Unused")
    plt.ylabel("Number of Loans")
    plt.xticks(rotation=90)
    plt.show()
```
